```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq
from scipy.interpolate import interp1d
```

```python
def Gauss(x, A, sigma, x0=0):
    """
    1d Gaussian function
    """
    return A/(np.sqrt(2*np.pi)*sigma) * np.exp(-(x - x0)**2/(2 * sigma**2))

x0 = np.arange(-5, 6, 1)
A0 = 1.0
sigma0 = 1.6/ 2.355
y0 = Gauss(x0, A0, sigma0)
x_shift = x0[:-1] + 0.5
y_shift = interp1d(x0, y0)(x_shift)
fit,_ = leastsq(lambda params, x, y: Gauss(x, params[0], params[1], params[2]) - y, [1.0, 1.5, 0], args = (x_shift, y_shift))
print fit
```

```python
fig, ax = plt.subplots()
xx = np.linspace(-5,6,500)
ax.plot(xx, Gauss(xx, A0, sigma0), color = 'r')
ax.plot(x0, y0, '--x', mew = 1.0, ms = 4, color = 'r')
ax.plot(x_shift, y_shift, 'o', ms = 4, color = 'g')
ax.plot(xx, Gauss(xx, fit[0], fit[1], fit[2]), color = 'g')
fig.savefig('fshift_problem.pdf')
```

```python
shiftList = np.linspace(-1, 1, 100)
fits = []
for shift in shiftList:
    x_shift = x0[1:-1] + shift
    y_shift = interp1d(x0, y0)(x_shift)
    fit,_ = leastsq(lambda params, x, y: Gauss(x, params[0], params[1], params[2]) - y, [1.0, 1.5, 0], args = (x_shift, y_shift))
    fits.append(fit)
    
sigma = [item[1] for item in fits]
plt.plot(shiftList, [sigma_i * 2.355 for sigma_i in sigma])
plt.xlabel('Fractional Shift')
plt.ylabel('FWHM fit result')
plt.ylim([1.6, 2.1])
plt.savefig('fshiftChangeFWHM.pdf')
```
