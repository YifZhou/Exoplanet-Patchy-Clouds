```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
f125DF = pd.read_csv('f125aperphot.dat', sep = '\s+', names = ['date-obs', 'time-obs', 'flux'], parse_dates = True)
f160DF = pd.read_csv('f160aperphot.dat', sep = '\s+', names = ['date-obs', 'time-obs', 'flux'], parse_dates = True)
```

```python
total = pd.concat([f125DF,f160DF], ignore_index = True)
total.sort(columns = ['date-obs', 'time-obs'], )
flt_file_info = pd.read_csv('../ABPIC-B_flt_file_info.csv', parse_dates = True)
```

* assemble CR removed file info

```python
crRemovedInfo = pd.DataFrame(columns = ['file name', 'orbit', 'filter', 'dither', 'No exposures', 'date-obs', 'time-obs'])
fn = []
orbit = []
filt = []
dither = []
Nexpos = []
date = []
time = []

for i, exposure in enumerate(set(flt_file_info['exposure'].values)):
    subdf = flt_file_info[flt_file_info['exposure'] == exposure]
    orbit.append(subdf['orbit'].values[0])
    dither.append(subdf['dither'].values[0])
    filt.append(subdf['filter'].values[0])
    fn.append('orbit_{0:02d}_dither_{1:02d}_{2}.fits'.format(orbit[i], dither[i], filt[i]))
    Nexpos.append(len(subdf))
    date.append(subdf['date-obs'].values[0])
    time.append(subdf['time-obs'].values[0])
crRemovedInfo = pd.DataFrame({'file name':fn, 'orbit':orbit, 'filter':filt, 'dither':dither, 'No exposures':Nexpos, 'date-obs':date, 'time-obs':time})
```

```python
crRemovedInfo.to_csv('ABPIC-B_CRremoved_fileInfo.csv', index = False)
```

```python
f125Info = crRemovedInfo[crRemovedInfo['filter'] == 'F125W']
f125Info.index = range(len(f125Info))
f125Info['flux'] = f125DF['flux']
f125Info['time'] = f125Info['date-obs'] + ' ' + f125Info['time-obs']+'Z'
f125Info['time'] = pd.to_datetime(f125Info['time'])
f125Info = f125Info.set_index(['time'])
f160Info = crRemovedInfo[crRemovedInfo['filter'] == 'F160W']
f160Info.index = range(len(f160Info))
f160Info['flux'] = f160DF['flux']
f160Info['time'] = f160Info['date-obs'] + ' ' + f160Info['time-obs']+'Z'
f160Info['time'] = pd.to_datetime(f160Info['time'])
f160Info = f160Info.set_index(['time'])
```

```python
time_interval = [(np.datetime64('2013-10-16T16:36:06Z'),np.datetime64('2013-10-16T17:29:24Z')),
                 (np.datetime64('2013-10-16 18:11:43Z'), np.datetime64('2013-10-16T19:05:02Z')),
                 (np.datetime64('2013-10-16 19:47:22Z'), np.datetime64('2013-10-16T20:40:40Z')),
                 (np.datetime64('2013-10-16 21:19:58Z'), np.datetime64('2013-10-16 22:16:17Z')),
                 (np.datetime64('2013-10-16 22:55:35Z'), np.datetime64('2013-10-16 23:51:55Z'))]
```

```python
%matplotlib inline
import matplotlib.transforms as mtransforms
fig = plt.figure()
ax = fig.add_subplot(111)
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.plot(f125Info.index, f125Info['flux'], marker = 'o', label = 'F125W', ms = 3)
ax.axhline(y = f125Info['flux'].mean(axis = 1), color = '0.8', ls = '--')
ax.plot(f160Info.index, f160Info['flux'], marker = 's', label = 'F160W', ms = 3)
ax.axhline(y = f160Info['flux'].mean(axis = 1), color = '0.8', ls = '--')
for time_int in time_interval:
    ax.fill_between(time_int, 0, 1, transform = trans, facecolor = '0.8', zorder = 5)
ax.set_ylim([8000, 8500])
ax.legend(loc = 'best')
ax.set_xlabel('UT')
ax.set_ylabel('Flux (counts/s)')
ax.set_title('Light Curve (Aperture = 5)')
fig.autofmt_xdate()
plt.savefig('fluxcurve_aper_05.pdf')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.plot(f125Info.index, f125Info['flux']/f125Info['flux'].mean(axis = 1), marker = 'o', label = 'F125W', ms = 3)
ax.plot(f160Info.index, f160Info['flux']/f160Info['flux'].mean(axis = 1), marker = 's', label = 'F160W', ms = 3)
for time_int in time_interval:
    ax.fill_between(time_int, 0, 1, transform = trans, facecolor = '0.8', zorder = 5)
ax.set_ylim([0.98, 1.015])
ax.legend(loc = 'best')
ax.set_xlabel('UT')
ax.set_ylabel('Normalized Flux')
ax.set_title('Light Curve (Aperture = 5)')
fig.autofmt_xdate()
plt.savefig('relativefluxcurve_aper_05.pdf')
```

```python
fileInfo = pd.read_csv('ABPIC-B_fileInfo.csv', parse_dates = {'dateTime': ['obs date', 'obs time']}, index_col='dateTime')
flux, fluxerr = np.loadtxt('aperphot_Dec21.dat', unpack = True)
fileInfo['flux'] = flux
fileInfo['fluxerr'] = fluxerr
print fileInfo.info()
fileInfo
F125DF = fileInfo[fileInfo['filter'] == 'F125W']
F160DF = fileInfo[fileInfo['filter'] == 'F160W']
```

```python
%matplotlib osx
fig = plt.figure()
ax = fig.add_subplot(111)
markers = ['$1$', '$2$', '$3$', '$4$']
markers1 = ['$-$', '+', '^', 's']
mean125 = F125DF['flux'].mean()
colors = ['r', 'g', 'b', 'm']
for dither in range(4):
    for expoSet in list(set(F125DF['exposure set'])):
        for orbit in range(7, 13):
            subframe = F125DF[(F125DF['dither'] == dither) & (F125DF['exposure set'] == expoSet) & (F125DF['orbit'] == orbit)]
            #ax.plot(subframe.index, subframe['flux'] + 168, linestyle = '--', marker = markers[dither], color = 'r', ms = 8)
            ax.plot(subframe.index, subframe['flux'], linestyle = '--', color = 'k', marker = markers[dither], mec = colors[dither], ms = 8)

for dither in range(4):
    for expoSet in list(set(F160DF['exposure set'])):
        for orbit in range(7, 13):
            subframe = F160DF[(F160DF['dither'] == dither) & (F160DF['exposure set'] == expoSet) & (F160DF['orbit'] == orbit)]
            ax.plot(subframe.index, subframe['flux'], linestyle = '-', color = 'purple', marker = markers1[dither], mec = colors[dither], mfc = 'none', ms = 8)



# ax.set_title('F125W')
ax.set_xlabel('UT')
ax.set_ylabel('Flux (Counts)')
ax.set_ylim([7950, 8500])
fig.autofmt_xdate()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
markers = ['$1$', '$2$', '$3$', '$4$']
colors = ['r', 'g', 'b', 'm']
mean160 = F160DF['flux'].mean()
for dither in range(4):
    for expoSet in list(set(F160DF['exposure set'])):
        for orbit in range(7, 13):
            subframe = F160DF[(F160DF['dither'] == dither) & (F160DF['exposure set'] == expoSet) & (F160DF['orbit'] == orbit)]
            ax.plot(subframe.index, subframe['flux']/mean160, linestyle = '--', marker = markers[dither], color = colors[dither], ms = 8)

ax.set_title('F160W')            
ax.set_xlabel('UT')
ax.set_ylabel('Relative Flux')
fig.autofmt_xdate()
```
