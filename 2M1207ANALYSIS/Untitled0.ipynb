```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DF = pd.read_csv('2M1207A_flux_Dec4.dat', sep = '\s+', names = ['File name', 'flux', 'x', 'y', 'obs Time', 'filter'], index_col= 'obs Time', parse_dates= True)
DF.sort_index()
```

```python
DF1 = DF[DF['filter'] == 'F125W'].sort_index()
DF2 = DF[DF['filter'] == 'F160W'].sort_index()
```

```python
%matplotlib osx
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(DF1.index, DF1['flux']/DF1['flux'].mean(axis = 1), '.', mew = 1)
ax.plot(DF2.index, DF2['flux']/DF2['flux'].mean(axis = 1), '.', mew = 1)
ax.plot()
fig.autofmt_xdate()
```
