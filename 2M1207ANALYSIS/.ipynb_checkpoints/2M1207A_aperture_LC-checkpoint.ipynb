```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib
```

```python
DF = pd.read_csv('2M1207A_flux_Dec4.dat', sep = '\s+', names = ['File name', 'flux', 'x', 'y', 'obs Time', 'filter'], index_col= 'obs Time', parse_dates= True)
```

```python
DF1 = DF[DF['filter'] == 'F125W'].sort_index()
DF2 = DF[DF['filter'] == 'F160W'].sort_index()
DF1dither = [1,1,1,2,2,2,3,3,3,4,4,
             1,1,1,2,2,2,3,3,3,4,4,
             1,1,1,2,2,2,3,3,3,4,4,4,
             1,1,1,2,2,2,3,3,3,4,4,4,
             1,1,1,2,2,2,3,3,3,4,4,4,
             1,1,1,2,2,2,3,3,3,4,4,4]

DF2dither = [1,1,1,2,2,3,3,3,4,4,
             1,1,1,2,2,3,3,3,4,4,
             1,1,1,2,2,3,3,3,4,4,4,
             1,1,1,2,2,3,3,3,4,4,4,
             1,1,1,2,2,3,3,3,4,4,4,
             1,1,1,2,2,3,3,3,4,4,4]

DF1['dither'] = DF1dither
DF2['dither'] = DF2dither
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
flux1 = DF1['flux'].mean(axis = 1)
for dither in [1, 2, 3, 4]:
    subFrame = DF1[DF1['dither'] == dither]
    ax.plot(subFrame.index, subFrame['flux']/flux1, 'o', mew = 0.1)
    
ax.plot(DF1.index, DF1['flux']/flux1, '--', color = '0.8')
ax.set_xlabel('UT')
ax.set_ylabel('Relative Flux')
ax.set_title('2M1207A F125W Aperture Photometry')
ax.set_ylim([0.97, 1.03])
fig.autofmt_xdate()
fig.savefig('2M1207A_F125W.pdf')

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
flux2 = DF2['flux'].mean(axis = 1)
for dither in [1, 2, 3, 4]:
    subFrame = DF2[DF2['dither'] == dither]
    ax1.plot(subFrame.index, subFrame['flux']/flux2, 'o', mew = 0.1)
    
ax1.plot(DF2.index, DF2['flux']/flux2, '--', color = '0.8')
ax1.set_ylim([0.97, 1.03])
ax1.set_xlabel('UT')
ax1.set_ylabel('Relative Flux')
ax1.set_title('2M1207A F160W Aperture Photometry')
fig1.autofmt_xdate()
fig1.savefig('2M1207A_F160W.pdf')
```
